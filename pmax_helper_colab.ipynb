## 4. ngrokの設定

ngrokを使ってFlaskサーバーを外部公開します。

```python
from pyngrok import ngrok

# ngrokトンネルを作成
public_url = ngrok.connect(5001)
print(f"\n🌐 バックエンドURL: {public_url}")
print(f"\n⚠️ このURLをフロントエンドの環境変数に設定してください:")
print(f"   VITE_API_URL={public_url}")
print(f"\n📝 フロントエンドで以下を実行:")
print(f"   echo 'VITE_API_URL={public_url}' > .env.local")
print(f"   npm run dev")
```
