## 1. 環境変数の設定

**重要：** Google ColabのSecrets機能を使ってAPIキーを設定してください。

### 手順:
1. 左サイドバーの **🔑 鍵アイコン（Secrets）** をクリック
2. 以下の3つのシークレットを追加:
   - **名前**: `FAL_KEY` → **値**: あなたのfal.ai APIキー
   - **名前**: `LAMBDA_SECRET_KEY` → **値**: ダミー値（例: `dummy_key`）
   - **名前**: `NGROK_AUTH_TOKEN` → **値**: あなたのngrok認証トークン
3. 各シークレットの「ノートブックからアクセスを許可」をONにする

### APIキーの取得方法:
- **fal.ai**: https://fal.ai/dashboard
- **ngrok**: https://dashboard.ngrok.com/get-started/your-authtoken （無料アカウント登録が必要）

```python
import os
from google.colab import userdata

# Google Colab Secretsから APIキーを取得
try:
    FAL_KEY = userdata.get('FAL_KEY')
    LAMBDA_SECRET_KEY = userdata.get('LAMBDA_SECRET_KEY')
    NGROK_AUTH_TOKEN = userdata.get('NGROK_AUTH_TOKEN')
    
    # 環境変数に設定
    os.environ['FAL_KEY'] = FAL_KEY
    os.environ['LAMBDA_SECRET_KEY'] = LAMBDA_SECRET_KEY
    
    print("✅ 環境変数を設定しました")
    print(f"   FAL_KEY: 設定済み ({FAL_KEY[:10]}...)")
    print(f"   LAMBDA_SECRET_KEY: 設定済み")
    print(f"   NGROK_AUTH_TOKEN: 設定済み ({NGROK_AUTH_TOKEN[:10]}...)")
except Exception as e:
    print("❌ エラー: Secretsが設定されていません")
    print("左サイドバーの🔑アイコンから以下のシークレットを追加してください:")
    print("  - FAL_KEY")
    print("  - LAMBDA_SECRET_KEY")
    print("  - NGROK_AUTH_TOKEN")
    print(f"\n詳細: {e}")
```
